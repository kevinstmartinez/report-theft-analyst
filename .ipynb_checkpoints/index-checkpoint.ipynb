```python
from pyspark import SparkContext, SparkConf
from pyspark.sql import SparkSession
from pyspark.sql import SQLContext
from pyspark.sql.types import StructType, StructField, IntegerType, StringType,FloatType, DateType
from pyspark.sql.functions import col, udf
from pyspark.sql import Window as W

from pyspark.sql import functions as F
import boto3
import os
import pandas as pd
from dotenv import load_dotenv
from datetime import datetime
from dateutil import parser
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

load_dotenv()


spark = SparkSession.builder\
.master("local[*]")\
.appName('reportehurtopormodalidades')\
.config('spark.jars.packages', 'org.apache.hadoop:hadoop-aws:3.3.4,org.apache.hadoop:hadoop-common:3.3.3')\
.config("spark.hadoop.fs.s3a.aws.credentials.provider", "org.apache.hadoop.fs.s3a.SimpleAWSCredentialsProvider")\
.config("spark.driver.memory", "18g")\
.getOrCreate()
sql_context = SQLContext(spark)




sc = spark.sparkContext

sc._jsc.hadoopConfiguration().set('fs.s3a.access.key', os.getenv('aws_access_key_id'))
sc._jsc.hadoopConfiguration().set('fs.s3a.secret.key', os.getenv('aws_secret_access_key'))
sc._jsc.hadoopConfiguration().set('fs.s3a.impl', 'org.apache.hadoop.fs.s3a.S3AFileSystem')

spark=SparkSession(sc)
```

```python

s3 = boto3.client(
    's3',
    aws_access_key_id = os.getenv('aws_access_key_id'),
    aws_secret_access_key = os.getenv('aws_secret_access_key'),
    region_name = os.getenv('region')
)

bucket = s3.list_buckets()['Buckets'][0]['Name']
files = s3.list_objects_v2(Bucket = bucket)

Key = files['Contents'][0]['Key']

path = f"s3a://{bucket}/{Key}"
```

```python

s3_df = spark.read.option('header', True).csv(path)


s3_df.printSchema()


def clean_values(value):
    return value.split(' ')[0]

def clean_date(value):
    return value.replace('/', '-')

clean_values_udf = udf(lambda z : clean_values(z), StringType())
sql_context.udf.register('clean_values', clean_values_udf)

clean_date_udf = udf(lambda z : clean_date(z), StringType())
sql_context.udf.register('clean_values', clean_date_udf)

s3_df = s3_df.select('DEPARTAMENTO', clean_values_udf('MUNICIPIO').alias('MUNICIPIO'), 'CODIGO DANE', 'ARMAS MEDIOS', clean_date_udf('FECHA HECHO').alias('FECHA HECHO'), 'GENERO', 'GRUPO ETARIO', 'TIPO DE HURTO', 'CANTIDAD')
```

```python
get_day =  udf (lambda x: x.split('-')[0])
get_month =  udf (lambda x: x.split('-')[1])
get_year =  udf (lambda x: x.split('-')[2])

df = s3_df.withColumn('DAY', get_day(col('FECHA HECHO')))\
     .withColumn('MONTH', get_month(col('FECHA HECHO')))\
     .withColumn('YEAR', get_year(col('FECHA HECHO')))
```

```python
df.printSchema()
```

```python
df_departments_with_most_theft = df.groupBy('DEPARTAMENTO').count().sort(col('count').desc()).limit(10)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.barplot(data = df_departments_with_most_theft.toPandas(), x='DEPARTAMENTO', y ='count')
```

```python
df_township_with_most_theft = df.groupBy('MUNICIPIO').count().sort(col('count').desc()).limit(10)
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.barplot(data = df_township_with_most_theft.toPandas(), x='MUNICIPIO', y ='count')
```

```python
df_department_detail_theft_1 = df.filter(col('DEPARTAMENTO') == 'ANTIOQUIA').groupBy('MUNICIPIO').count().sort(col('count').desc()).limit(10)
```

```python
df_department_detail_theft_1.show(truncate=False)
```

```python
df_weapon_most_used = df.groupBy('ARMAS MEDIOS').count().sort(col('count').desc()).limit(5)
```

```python
df_weapon_most_used.show()
```

```python
df_weapon_most_used.createOrReplaceTempView('weapon')
```

```python
df_weapon = sql_context.sql(
"""
SELECT `ARMAS MEDIOS`, ROUND((count /
(SELECT sum(count) FROM weapon
GROUP BY 'ARMAS MEDIOS') * 100)) AS percent  FROM weapon
"""
)
```

```python
df_weapon.show(truncate=False)
```

```python
colors = sns.color_palette('pastel')[0:5]
labels = ['LLAVE MAESTRA','ARMA DE FUEGO','SIN EMPLEO DE ARMAS','ARMA BLANCA / CORTOPUNZANTE','NO REPORTADO']
plt.pie(data=df_weapon.toPandas(), x='percent', labels = labels, colors = colors, autopct='%.0f%%')
plt.show()
```
